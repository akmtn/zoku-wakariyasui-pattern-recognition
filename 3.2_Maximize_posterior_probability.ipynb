3.2　事後確率最大化

```python
import numpy as np
import matplotlib.pyplot as plt
%precision 3
import seaborn
```

```python
#既知のパラメータ
pi_1 = 0.1
pi_2 = 0.4
pi_3 = 0.5
th_1 = 0.8
th_2 = 0.6
th_3 = 0.3
```

```python
#事後確率を求める
def out_postPr(n,r):
    #margiPr : 周辺確率
    margiPr = pi_1 * th_1**r * (1-th_1)**(n-r) + pi_2 * th_2**r * (1-th_2)**(n-r) + pi_3 * th_3**r * (1-th_3)**(n-r)
    
    return [pi_1 * th_1**r * (1-th_1)**(n-r) / margiPr, pi_2 * th_2**r * (1-th_2)**(n-r) / margiPr, pi_3 * th_3**r * (1-th_3)**(n-r) / margiPr]
```

```python
#「観測結果と事後確率の関係(n=10)」をプロット
postPr = []
for i in range(11):
    postPr.append(out_postPr(10,i))

postPr = np.array(postPr)

plt.plot(postPr[:,0], label='w1')
plt.plot(postPr[:,1], label='w2')
plt.plot(postPr[:,2], label='w3')
plt.legend()
plt.show()
```

3.3　事前確率の効果

```python
#「観測結果と事後確率の関係(n=100)」をプロット
postPr = []
for i in range(101):
    postPr.append(out_postPr(100,i))

postPr = np.array(postPr)

plt.plot(postPr[:,0], label='w1')
plt.plot(postPr[:,1], label='w2')
plt.plot(postPr[:,2], label='w3')
plt.legend()
plt.show()
```
