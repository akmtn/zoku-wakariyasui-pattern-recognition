4.7 ベイズ推定の実験

```python
import numpy as np
import matplotlib.pyplot as plt
%precision 3
%matplotlib inline
import seaborn
from scipy.special import comb
```

```python
def out_postPr(n,r,th):
    return  (n+1) * comb(n, r) * th**r * (1-th)**(n-r)
```

```python
def plot(n,r):
    plt.figure(figsize=(4,3))
    th = np.linspace(0,1,100)
    plt.plot(th, out_postPr(n,r,th))
    plt.xlim(0,1)
    plt.xlabel('param_th')
    plt.ylim(0,4)
    plt.ylabel('Postriori Probability')
    plt.title('param_th by Bayse Estimation')
    plt.show()
```

図4.3「パラメータ推定の経過」の再現  
観測結果xからnとrの値を求める．

```python
#図４．３「パラメータ推定の経過」
x_10 = 'HHHHTHHTHT'
n = 0
r = 0 
for i, xi in enumerate(x_10):
    if xi == 'H':
        n += 1
        r += 1
        plot(n,r)
        
    elif xi == 'T':
        n += 1
        plot(n,r)
```

n=100の時，n=1000の時

```python
#観測データの生成
x_1000 = ''
for i in np.random.rand(1000):
    if i <= 0.8:
        x_1000 += 'H'
    else:
        x_1000 += 'T'
x_100 = x_1000[:100]
x_10 = x_1000[:10]
```

```python
#図４．４　「ベイズ推定によるパラメータ推定の実験」
th = np.linspace(0,1,100)
plt.plot(th, out_postPr(0,0,th),label='n = 0',linestyle='--')

for x_n in [x_10, x_100, x_1000]:
    n = 0
    r = 0 
    
    for i, xi in enumerate(x_n):
        if xi == 'H':
            n += 1
            r += 1   
        elif xi == 'T':
            n += 1

    
    plt.plot(th, out_postPr(n,r,th),label='n = '+str(i+1))
plt.xlabel('param_th')
plt.ylabel('Postriori Probability')
plt.title('param_th by Bayse Estimation')
plt.legend(loc='best')
plt.show()
```

```python
#図４．５「事前分布の影響（θ=0.2にピーク値を持つ事前分布）」
th = np.linspace(0,1,200)
plt.plot(th, out_postPr(10,2,th),label='n = 0',linestyle='--')

for x_n in [x_10, x_100, x_1000]:
    n = 10
    r = 2 
    
    for i, xi in enumerate(x_n):
        if xi == 'H':
            n += 1
            r += 1   
        elif xi == 'T':
            n += 1

    
    plt.plot(th, out_postPr(n,r,th),label='n = '+str(i+1))
plt.xlabel('param_th')
plt.ylabel('Postriori Probability')
plt.title('param_th by Bayse Estimation')
plt.legend(loc='best')
plt.show()
```

```python
#図４．６「事前分布の影響（θ=0.6にピーク値を持つ事前分布）」
th = np.linspace(0,1,200)
plt.plot(th, out_postPr(50,30,th),label='n = 0',linestyle='--')

for x_n in [x_10, x_100, x_1000]:
    n = 50
    r = 30 
    
    for i, xi in enumerate(x_n):
        if xi == 'H':
            n += 1
            r += 1   
        elif xi == 'T':
            n += 1

    
    plt.plot(th, out_postPr(n,r,th),label='n = '+str(i+1))
plt.xlabel('param_th')
plt.ylabel('Postriori Probability')
plt.title('param_th by Bayse Estimation')
plt.legend(loc='best')
plt.show()
```
