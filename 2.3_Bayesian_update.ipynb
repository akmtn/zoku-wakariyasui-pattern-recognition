第２章　事前確率と事後確率  
2.3 ベイズの更新の実験

π1 = 0.1,

θ1 = 0.8,

```python
#事前確率から事後確率を求める
def out_postPr(priPr,x):
    if x == 'H':
        #分母の周辺確率を計算
        margiPr = priPr[0]*0.8 + priPr[1]*0.6 + priPr[2]*0.3 
        
        #ωごとに事後確率を計算
        pr_x_w1 = 0.8 * priPr[0] / margiPr
        pr_x_w2 = 0.6 * priPr[1] / margiPr
        pr_x_w3 = 0.3 * priPr[2] / margiPr
        
        return [pr_x_w1, pr_x_w2, pr_x_w3]
        
    elif x == 'T':
        #分母の周辺確率を計算
        margiPr = priPr[0]*0.2 + priPr[1]*0.4 + priPr[2]*0.7 
        
        #ωごとに事後確率を計算
        pr_x_w1 = 0.2 * priPr[0] / margiPr
        pr_x_w2 = 0.4 * priPr[1] / margiPr
        pr_x_w3 = 0.7 * priPr[2] / margiPr
        
        return [pr_x_w1, pr_x_w2, pr_x_w3]
```

```python
postPr = out_postPr([0.1, 0.4, 0.5],'H')
```

10回の結果を入力した後の事後確率を求める

```python
postPr = [[0.1,0.4,0.5]]
x_10 = 'HHHHTHHTHT'
for i, x in enumerate(x_10):
    postPr.append(out_postPr(postPr[i],x))
    
```

```python
postPr = np.array(postPr)
postPr[:,1]
```

```python
plt.plot(postPr[:,0],label='w1')
plt.plot(postPr[:,1],label='w2')
plt.plot(postPr[:,2],label='w3')
plt.legend(loc='upper left')
plt.title('Transition of Posteriori Probability')
plt.show()
```

```python
x_100 = ''
for i in np.random.rand(100):
    if i <= 0.8:
        x_100 += 'H'
    else:
        x_100 += 'T'
print(x_100)
```

```python
postPr = [[0.1,0.4,0.5]] #postPr[0]は事前確率πi
for i, x in enumerate(x_100):
    postPr.append(out_postPr(postPr[i],x))
    
postPr = np.array(postPr)

plt.plot(postPr[:,0],label='w1')
plt.plot(postPr[:,1],label='w2')
plt.plot(postPr[:,2],label='w3')
plt.legend(loc='best')
plt.title('Transition of Posteriori Probability')
plt.ylabel('Posteriori Probability')
plt.show()
```
